```python
from PIL import Image, ImageDraw, ImageFont, ImageOps
from pathlib import Path

# ระบุโฟลเดอร์ที่ต้องการอ่านไฟล์ทั้งหมด
output_folder = Path("data/raw_images/YouareMyspecial")
folder_path = Path("font/GoogleFont")


if not folder_path.exists():
    print(f"Error: Font folder does not exist at {folder_path}")
    exit(1)
output_folder.mkdir(parents=True, exist_ok=True)

# เริ่มสร้างภาพ
for font_number, font_file in enumerate(folder_path.rglob("*.ttf"), start=1):
    if font_file.is_file() and font_file.suffix.lower() == '.ttf':
        for char in "+-*().=/\|":#ขีดกะลบตัวเดียวกัน(-) ทับกะหารตัวเดียวกัน(/)
            try:
                if(char == "-" ):
                    font = ImageFont.truetype(str(font_file), 66)
                elif(char == "*" ):
                    font = ImageFont.truetype(str(font_file), 66)
                elif(char == "+" ):
                    font = ImageFont.truetype(str(font_file), 60)
                elif(char == "=" ):
                    font = ImageFont.truetype(str(font_file), 60)
                elif(char == "(" or char == ")" ):
                    font = ImageFont.truetype(str(font_file), 32)
                elif(char == "."  ):
                    font = ImageFont.truetype(str(font_file), 105)
                elif(char == "/"  or char == "|" or char == "\\" ):
                    font = ImageFont.truetype(str(font_file), 32)
                else:
                    font = ImageFont.truetype(str(font_file), 46)
            except Exception as e:
                print(f"Skipping font {font_file}: {e}")
                continue

            # สร้างโฟลเดอร์สำหรับตัวเลข
            if char == "*":
                char_folder = output_folder / "multiply"
            elif char == ".":
                char_folder = output_folder / "dot"
            elif char == "/":
                char_folder = output_folder / "slash"
            elif char == "\\":
                char_folder = output_folder / "backslash"
            elif char == "|":
                char_folder = output_folder / "pipe"
            else:
                char_folder = output_folder / f"{char}"
            
            char_folder.mkdir(parents=True, exist_ok=True)

            # คำนวณ Ascent และ Descent ของฟอนต์
            ascent, descent = font.getmetrics()

            # คำนวณ Bounding Box
            dummy_image = Image.new("L", (1, 1))
            draw = ImageDraw.Draw(dummy_image)
            bbox = draw.textbbox((0, 0), char, font=font)

            text_width = bbox[2] - bbox[0]
            text_height = bbox[3] - bbox[1]

            # สร้างภาพพอดีตัวอักษร
            image = Image.new('L', (text_width, ascent + descent), color=0)
            draw = ImageDraw.Draw(image)

            # ปรับตำแหน่ง Y ทั่วไป
            y_offset = ascent - bbox[3]
            x_offset = 0

            # ปรับตำแหน่ง y_offset เฉพาะจุด
            # +ลง -ขึ้น
            if(char == "-"):
                if(1 <= font_number <= 26):
                    y_offset+=15
                elif(27 <= font_number <= 36):
                    y_offset-=4
                elif(37 <= font_number <= 49):
                    y_offset-=6
                elif(51 <= font_number <= 63):
                    y_offset+=3
                elif(font_number == 64):
                    y_offset-=8
                elif(65 <= font_number <= 80):
                    y_offset-=5
                elif(81 <= font_number <= 96):
                    y_offset-=3
                elif(67 <= font_number <= 110):
                    y_offset-=6
                elif(111 <= font_number <= 122):
                    y_offset-=8
                elif(123 <= font_number <= 134):
                    y_offset-=2
                elif(135 <= font_number <= 140):
                    y_offset-=0
                elif(141 <= font_number <= 152):
                    y_offset-=2
                elif(153 <= font_number <= 158):
                    y_offset-=4
                elif(159 <= font_number <= 170):
                    y_offset-=10
                elif(171 <= font_number <= 207):
                    y_offset-=1
                elif(208 <= font_number <= 216):
                    y_offset-=10
                elif(217 <= font_number <= 253):
                    y_offset+=2
                elif(254 <= font_number <= 259):
                    y_offset-=0
                elif(260 <= font_number <= 277):
                    y_offset-=3
                elif(278 <= font_number <= 293):
                    y_offset-=8
                elif(294 <= font_number <= 311):
                    y_offset-=1
                elif(312 <= font_number <= 315):
                    y_offset-=2
                elif(316 <= font_number <= 333):
                    y_offset-=2
            
            if(char == "*"):
                if(1 <= font_number <= 2 or 9<= font_number <= 14 or font_number ==7):
                    y_offset+=27
                elif(3<= font_number <= 6 or font_number ==8 or 15<= font_number <= 26 ):
                    y_offset+=32
                elif(27 <= font_number <= 36):
                    y_offset+=15
                elif(37 <= font_number <= 49):
                    y_offset+=9
                elif(font_number == 50 or font_number == 57 or font_number == 55 or font_number == 62 ):
                    y_offset+=12
                elif(51 <= font_number <= 54 or font_number == 56 or font_number == 63 or 58 <= font_number <= 61 ):
                    y_offset+=18
                elif(64 <= font_number <= 80):
                    y_offset+=8
                elif(81 <= font_number <= 110):
                    y_offset+=8
                elif(111 <= font_number <= 122):
                    y_offset+=12
                elif(123 <= font_number <= 134):
                    y_offset+=22
                elif(135 <= font_number <= 140):
                    y_offset+=15
                elif(141 <= font_number <= 152):
                    y_offset+=6
                elif(153 <= font_number <= 158):
                    y_offset+=12
                elif(159 <= font_number <= 170):
                    y_offset+=6
                elif(171 <= font_number <= 207):
                    y_offset+=15
                elif(208 <= font_number <= 216):
                    y_offset+=7
                elif(217 <= font_number <= 253):
                    y_offset+=20
                elif(254 <= font_number <= 259):
                    y_offset+=15
                elif(260 <= font_number <= 277):
                    y_offset+=15
                elif(278 <= font_number <= 293):
                    y_offset+=8
                elif(294 <= font_number <= 311):
                    y_offset+=12
                elif(312 <= font_number <= 315):
                    y_offset+=15
                elif(316 <= font_number <= 333):
                    y_offset+=12
            if(char == "(" or char == ")"):
                if(1 <= font_number <= 26):
                    y_offset+=22
                elif(27 <= font_number <= 36):
                    y_offset+=6
                elif(37 <= font_number <= 49 or font_number == 64):
                    y_offset+=3
                elif(50 <= font_number <= 63):
                    y_offset+=5
                elif(65 <= font_number <= 80):
                    y_offset+=2
                elif(81 <= font_number <= 110):
                    y_offset+=2
                elif(111 <= font_number <= 122):
                    y_offset+=4
                elif(123 <= font_number <= 134):
                    y_offset+=4
                elif(135 <= font_number <= 140):
                    y_offset+=4
                elif(141 <= font_number <= 152):
                    y_offset+=3
                elif(153 <= font_number <= 158):
                    y_offset+=4
                elif(159 <= font_number <= 170):
                    y_offset-=0
                elif(171 <= font_number <= 207):
                    y_offset+=6
                elif(208 <= font_number <= 216):
                    y_offset-=0
                elif(217 <= font_number <= 253):
                    y_offset+=6
                elif(254 <= font_number <= 259):
                    y_offset+=3
                elif(260 <= font_number <= 277):
                    y_offset+=5
                elif(278 <= font_number <= 293):
                    y_offset-=0
                elif(294 <= font_number <= 311):
                    y_offset+=5
                elif(312 <= font_number <= 315):
                    y_offset+=2
                elif(316 <= font_number <= 333):
                    y_offset+=2
            if(char == "."):
                if(1 <= font_number <= 26):
                    y_offset-=15
                    x_offset+=2
                elif(27 <= font_number <= 36):
                    y_offset-=30
                    x_offset+=2
                elif(37 <= font_number <= 49):
                    y_offset-=30
                    x_offset+=2
                elif(50 <= font_number <= 63):
                    y_offset-=20
                elif(64 <= font_number <= 80):
                    y_offset-=33
                    x_offset+=2
                elif(97<= font_number <= 110 or 94<= font_number <= 95 or 90 <= font_number <= 92 or 87 <= font_number <= 88 or font_number == 81 or font_number == 83 or  font_number == 85 ):
                    y_offset-=30
                    x_offset+=2
                elif( font_number == 82 or font_number == 84 or font_number == 86 or font_number == 89 or font_number == 93 or font_number == 96 ):
                    y_offset-=30
                    x_offset+=7
                elif(111 <= font_number <= 122):
                    y_offset-=35
                    x_offset+=2
                elif(123 <= font_number <= 134):
                    y_offset-=30
                    x_offset+=2
                elif(135 <= font_number <= 140):
                    y_offset-=25
                    x_offset+=0
                elif(141 <= font_number <= 152):
                    y_offset-=35
                    x_offset+=2
                elif(153 <= font_number <= 158):
                    y_offset-=30
                    x_offset+=2
                elif(159 <= font_number <= 170):
                    y_offset-=40
                    x_offset+=2
                elif(171 <= font_number <= 207):
                    y_offset-=25
                    x_offset+=2
                elif(208 <= font_number <= 216):
                    y_offset-=37
                    x_offset+=2
                elif(217 <= font_number <= 253):
                    y_offset-=20
                    x_offset+=2
                elif(254 <= font_number <= 259):
                    y_offset-=30
                    x_offset+=2
                elif(260 <= font_number <= 277):
                    y_offset-=30
                    x_offset+=2
                elif(278 <= font_number <= 291):
                    y_offset-=35
                    x_offset+=2
                elif(292 <= font_number <= 293):
                    y_offset-=40
                    x_offset+=2
                elif(294 <= font_number <= 311):
                    y_offset-=25
                    x_offset+=2
                elif(312 <= font_number <= 315):
                    y_offset-=30
                    x_offset+=2
                elif(316 <= font_number <= 321):
                    y_offset-=27
                    x_offset+=2
                elif(322 <= font_number <= 333):
                    y_offset-=30
                    x_offset+=2
            if(char == "="):
                if(1 <= font_number <= 26):
                    y_offset+=15
                elif(27 <= font_number <= 36):
                    y_offset-=2
                elif(37 <= font_number <= 49):
                    y_offset-=3
                elif(64 <= font_number <= 80):
                    y_offset-=4
                elif(81 <= font_number <= 110):
                    y_offset-=5
                elif(111 <= font_number <= 122):
                    y_offset-=5
                elif(123 <= font_number <= 134):
                    y_offset-=2
                elif(135 <= font_number <= 140):
                    y_offset-=2
                elif(141 <= font_number <= 152):
                    y_offset-=4
                elif(153 <= font_number <= 158):
                    y_offset-=4
                elif(159 <= font_number <= 170):
                    y_offset-=7
                elif(171 <= font_number <= 207):
                    y_offset-=1
                elif(208 <= font_number <= 216):
                    y_offset-=7
                elif(217 <= font_number <= 253):
                    y_offset+=2
                elif(254 <= font_number <= 259):
                    y_offset-=3
                elif(260 <= font_number <= 277):
                    y_offset-=3
                elif(278 <= font_number <= 293):
                    y_offset-=5
                elif(294 <= font_number <= 311):
                    y_offset-=1
                elif(312 <= font_number <= 315):
                    y_offset-=2
                elif(316 <= font_number <= 333):
                    y_offset-=2
            if(char == "+"):
                if(1 <= font_number <= 26):
                    y_offset+=15
                elif(27 <= font_number <= 36):
                    y_offset-=2
                elif(37 <= font_number <= 49):
                    y_offset-=3
                elif(64 <= font_number <= 80):
                    y_offset-=8
                elif(81 <= font_number <= 86):
                    y_offset-=3
                elif(87 <= font_number <= 110):
                    y_offset-=5
                elif(111 <= font_number <= 122):
                    y_offset-=5
                elif(123 <= font_number <= 134):
                    y_offset-=2
                elif(135 <= font_number <= 140):
                    y_offset-=2
                elif(141 <= font_number <= 152):
                    y_offset-=4
                elif(153 <= font_number <= 158):
                    y_offset-=4
                elif(159 <= font_number <= 170):
                    y_offset-=8
                elif(171 <= font_number <= 207):
                    y_offset+=3
                elif(208 <= font_number <= 216):
                    y_offset-=7
                elif(217 <= font_number <= 253):
                    y_offset+=2
                elif(254 <= font_number <= 259):
                    y_offset-=3
                elif(260 <= font_number <= 277):
                    y_offset-=3
                elif(278 <= font_number <= 293):
                    y_offset-=5
                elif(294 <= font_number <= 311):
                    y_offset-=1
                elif(312 <= font_number <= 315):
                    y_offset-=2
                elif(316 <= font_number <= 333):
                    y_offset-=2
            if(char == "/"  or char == "|" or char == "\\"):
                if(1 <= font_number <= 26):
                    y_offset+=1
                elif(27 <= font_number <= 36):
                    y_offset-=1
                elif(37 <= font_number <= 49):
                    y_offset+=1
                elif(50<= font_number <= 63):
                    y_offset+=4
                elif(64 <= font_number <= 80):
                    y_offset-=1
                elif(81 <= font_number <= 110):
                    y_offset+=1
                elif(111 <= font_number <= 122):
                    y_offset+=2
                elif(123 <= font_number <= 134):
                    y_offset+=2
                elif(135 <= font_number <= 140):
                    y_offset+=4
                elif(141 <= font_number <= 152):
                    y_offset+=2
                elif(153 <= font_number <= 158):
                    y_offset+=2
                elif(159 <= font_number <= 170):
                    y_offset-=2
                elif(171 <= font_number <= 207):
                    y_offset+=3
                elif(208 <= font_number <= 216):
                    y_offset-=1
                elif(217 <= font_number <= 253):
                    y_offset+=6
                elif(254 <= font_number <= 259):
                    y_offset+=1
                elif(260 <= font_number <= 277):
                    y_offset+=2
                elif(278 <= font_number <= 293):
                    y_offset+=0
                elif(294 <= font_number <= 311):
                    y_offset-=0
                elif(312 <= font_number <= 315):
                    y_offset+=3
                elif(316 <= font_number <= 333):
                    y_offset+=1
            
            else:
                if(1 <= font_number <= 26):
                    y_offset-=19
                elif(27 <= font_number <= 36):
                    y_offset-=2
                elif(37 <= font_number <= 49):
                    y_offset-=3
                elif(64 <= font_number <= 80):
                    y_offset-=4
                elif(81 <= font_number <= 110):
                    y_offset-=3
                elif(111 <= font_number <= 122):
                    y_offset-=5
                elif(123 <= font_number <= 134):
                    y_offset-=2
                elif(135 <= font_number <= 140):
                    y_offset-=2
                elif(141 <= font_number <= 152):
                    y_offset-=4
                elif(153 <= font_number <= 158):
                    y_offset-=4
                elif(159 <= font_number <= 170):
                    y_offset-=7
                elif(171 <= font_number <= 207):
                    y_offset-=1
                elif(208 <= font_number <= 216):
                    y_offset-=7
                elif(217 <= font_number <= 253):
                    y_offset+=2
                elif(254 <= font_number <= 259):
                    y_offset-=3
                elif(260 <= font_number <= 277):
                    y_offset-=3
                elif(278 <= font_number <= 293):
                    y_offset-=5
                elif(294 <= font_number <= 311):
                    y_offset-=1
                elif(312 <= font_number <= 315):
                    y_offset-=2
                elif(316 <= font_number <= 333):
                    y_offset-=2

            
            draw.text((x_offset, y_offset), char, fill=255, font=font)

            # เพิ่ม Padding เพื่อให้ได้ขนาด 32x32
            desired_size = 32
            delta_w = desired_size - image.width
            delta_h = desired_size - image.height
            padding = (delta_w // 2, delta_h // 2, delta_w - (delta_w // 2), delta_h - (delta_h // 2))
            padded_image = ImageOps.expand(image, padding, fill=0)

            # บันทึกภาพ
            sanitized_char = char.replace("*", "multiply").replace(".", "dot").replace("/", "slash").replace("\\", "backslash").replace("|", "pipe")
            output_path = char_folder / f"{sanitized_char}_{font_number}.png"
            padded_image.save(output_path)
            print(f"Saved: {output_path}")

```
