```python
import easyocr
import cv2
import numpy as np
from matplotlib import pyplot as plt
from PIL import ImageFont, ImageDraw, Image
```

```python
reader = easyocr.Reader(['en', 'th'], gpu=True)
```

```python
image_path = 'transcript/pun_20240817_0001.jpg'
image = cv2.imread(image_path)

# ตรวจจับข้อความ
results = reader.readtext(image_path)
```

```python
# แปลงภาพจาก cv2 เป็น PIL เพื่อใช้ฟอนต์ภาษาไทย
image_pil = Image.fromarray(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
draw = ImageDraw.Draw(image_pil)
font_path = 'angsana-new.ttf'  # ระบุที่ตั้งของฟอนต์ภาษาไทย
font = ImageFont.truetype(font_path, 20)

# วาด bounding box และข้อความบนภาพ
for (bbox, text, prob) in results:
    # bbox เป็น list ของจุด 4 จุดของกรอบ
    (top_left, top_right, bottom_right, bottom_left) = bbox
    top_left = tuple([int(x) for x in top_left])
    bottom_right = tuple([int(x) for x in bottom_right])

    # วาด rectangle บนภาพ
    draw.rectangle([top_left, bottom_right], outline=(0, 255, 0), width=2)

    # ใส่ข้อความที่ถูกตรวจจับด้วยฟอนต์ภาษาไทย
    draw.text((top_left[0], top_left[1] - 20), text, font=font, fill="red")

# แปลงภาพกลับเป็น cv2 และแสดงภาพ
image = cv2.cvtColor(np.array(image_pil), cv2.COLOR_RGB2BGR)
output_path = 'output_image.jpg'
cv2.imwrite(output_path, image)

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
