```python
def add_leading_zero(text):
    # ถ้าความยาวของ text น้อยกว่า 3 เติม 0 ด้านหน้า
    if len(text) < 3:
        return text.zfill(3)  # zfill จะเติม 0 จนความยาวครบ 3
    return text  # ถ้าความยาวมากกว่าหรือเท่ากับ 3 ไม่ต้องเติม
```
