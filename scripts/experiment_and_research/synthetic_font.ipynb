```python
from PIL import Image, ImageDraw, ImageFont

# สร้างภาพเปล่า (32x32)
image_size = (32, 32)
image = Image.new('L', image_size, color=0)  # สีพื้นหลังดำ (Grayscale)

# โหลดฟอนต์
font_file = "../../Font/GoogleFont/Sarabun/Sarabun-Bold.ttf"  # เปลี่ยนเป็น path ของฟอนต์จริง
font = ImageFont.truetype(str(font_file), 28)

# สร้าง ImageDraw สำหรับวาดตัวอักษร
draw = ImageDraw.Draw(image)

# ข้อความที่ต้องการวาด
text = "A"

# คำนวณ Bounding Box ของข้อความ
bbox = draw.textbbox((0, 0), text, font=font)  # (left, top, right, bottom)
text_width, text_height = bbox[2] - bbox[0], bbox[3] - bbox[1]

# คำนวณตำแหน่งให้อยู่กึ่งกลางแนวแกน X และ Y
x_position = (image_size[0] - text_width) // 2
y_position = (image_size[1] - text_height*2) // 2
# วาดตัวอักษรลงในภาพ
draw.text((x_position, y_position), text, fill=255, font=font)  # สีตัวอักษรขาว (255)

# แสดงภาพ
image.show()

# บันทึกภาพ
image.save("output_centered.png")
```

```python
from PIL import Image, ImageDraw, ImageFont

# โหลดฟอนต์
font_file = "../../Font/GoogleFont/Sarabun/Sarabun-Bold.ttf"  # เปลี่ยนเป็น path ของฟอนต์จริง
font = ImageFont.truetype(str(font_file), 28)

# ข้อความที่ต้องการวาด
text = "j"

# คำนวณ Ascent และ Descent ของฟอนต์
ascent, descent = font.getmetrics()

# คำนวณ Bounding Box ของข้อความ
dummy_image = Image.new('L', (1, 1))  # สร้างภาพชั่วคราวเพื่อคำนวณขนาด
draw = ImageDraw.Draw(dummy_image)
bbox = draw.textbbox((0, 0), text, font=font)  # (left, top, right, bottom)

# คำนวณขนาดภาพ
text_width = bbox[2] - bbox[0]
text_height = bbox[3] - bbox[1]

# สร้างภาพที่มีขนาดพอดีกับตัวอักษร + Adjust Y
image = Image.new('L', (text_width, ascent + descent), color=0)  # รวม Ascent + Descent
draw = ImageDraw.Draw(image)

# วาดตัวอักษรโดยปรับแกน Y ให้สมดุล
y_offset = ascent - bbox[3]  # ปรับตำแหน่ง Y ให้สมดุล
draw.text((0, y_offset), text, fill=255, font=font)

# แสดงผล
image.show()

# บันทึกภาพ
image.save("output_y_centered.png")
```

```python
from PIL import Image, ImageDraw, ImageFont, ImageOps

# โหลดฟอนต์
font_file = "../../Font/GoogleFont/Sarabun/Sarabun-Bold.ttf"  # เปลี่ยนเป็น path ของฟอนต์จริง
font = ImageFont.truetype(str(font_file), 28)

# ข้อความที่ต้องการวาด
text = "Y"

# คำนวณ Ascent และ Descent ของฟอนต์
ascent, descent = font.getmetrics()

# สร้างภาพชั่วคราวเพื่อคำนวณ Bounding Box ของตัวอักษร
dummy_image = Image.new('L', (1, 1))  # ภาพขนาดเล็กเพื่อคำนวณขนาด
draw = ImageDraw.Draw(dummy_image)
bbox = draw.textbbox((0, 0), text, font=font)  # (left, top, right, bottom)

# คำนวณขนาดของตัวอักษร
text_width = bbox[2] - bbox[0]
text_height = bbox[3] - bbox[1]

# สร้างภาพพอดีกับตัวอักษร
image = Image.new('L', (text_width, ascent + descent), color=0)  # สีดำพื้นหลัง
draw = ImageDraw.Draw(image)

# ปรับตำแหน่ง Y เพื่อให้อยู่ตรงกลางโดยใช้ ascent
y_offset = ascent - bbox[3]
draw.text((0, y_offset), text, fill=255, font=font)  # สีตัวอักษรเป็นสีขาว

# เพิ่ม Padding เพื่อให้ได้ขนาด 32x32
desired_size = 32
delta_w = desired_size - image.width
delta_h = desired_size - image.height
padding = (delta_w // 2, delta_h // 2, delta_w - (delta_w // 2), delta_h - (delta_h // 2))

# เพิ่ม Padding รอบภาพ
padded_image = ImageOps.expand(image, padding, fill=0)  # สี Padding เป็นสีดำ

# แสดงภาพ
padded_image.show()

# บันทึกภาพ
padded_image.save("output_32x32.png")
```
